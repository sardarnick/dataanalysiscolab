```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
tips = sns.load_dataset("tips")
print(tips.head())

tips.head(4)
plt.scatter(tips['total_bill'], tips['day'])
plt.xlabel('Bill')
plt.ylabel('Day of week')
plt.title('Bill by day of week')
plt.show()

sns.boxplot(data=tips, x='tip', y='sex')
plt.title('Employee Satisfaction by Department')
plt.xticks(rotation=45)
plt.show()



```
