```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np # Import numpy library and alias it as np
data = pd.read_csv('Business_Management_Dataset.csv')

#training hours and productivity scatter plot

sns.scatterplot(data=data, x='Training_Hours', y='Productivity', hue='Department')
plt.title('Training Hours vs Productivity')
plt.show()

#Employee_Satisfaction across different Departments boxplots
sns.boxplot(data=data, x='Department', y='Employee_Satisfaction')
plt.title('Employee Satisfaction by Department')
plt.xticks(rotation=45)
plt.show()
#Pair Plot
#sns.pairplot(data, hue='Department')
#plt.show()

#To show average Annual_Sales by Company_Type: bar
sns.barplot(data=data, x='Company_Type', y='Annual_Sales', estimator=sum)
plt.title('Total Annual Sales by Company Type')
plt.show()

#Heatmap To visualize correlations between numeric variables:
numeric_data = data.select_dtypes(include=np.number)
plt.figure(figsize=(10, 8))
sns.heatmap(numeric_data.corr(), annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
